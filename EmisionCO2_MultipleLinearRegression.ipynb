```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
```

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv"
```

```python
df=pd.read_csv(url)
df.head()
```

```python
cdf=df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
msk=np.random.rand(len(df))<0.8
train=cdf[msk]
test=cdf[~msk]

train_x=np.asarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
train_y=np.asarray(train[['CO2EMISSIONS']])
test_x=np.asarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
test_y=np.asarray(test[['CO2EMISSIONS']])
```

```python
from sklearn import linear_model

regr=linear_model.LinearRegression()

regr.fit(train_x,train_y)

print('Coeficientes: ',regr.coef_)
print('Intersección: ',regr.intercept_)
```

```python
prediccion=regr.predict(test_x)


RSS=np.mean((prediccion-test_y)**2)
VarScore=regr.score(test_x,test_y)

print(f"RSS:{RSS}")
print(f"VAr score:{VarScore}")
```
